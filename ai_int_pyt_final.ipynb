## 5

a. drop column > 50 % data missing  

```python
# kolom 'company' akan dihapus karena  melebihi 50% data kosong
del df_base['company']

print(df_base.info())
```

b. fill the missing value in 'agent' with mean

```python
df_base['agent'] = df_base['agent'].fillna(df_base['agent'].mean())
# df_base.head(10)
df_base.info()
```

c. fiil the value in 'country' and 'children'

```python
df_base['country'] = df_base['country'].fillna(df_base['country'].mode()[0])
df_base['children'] = df_base['children'].fillna(df_base['children'].mode()[0])
df_base.info()
```

```python
# Visualize the arrival_date_month to see what month most visitor come
import matplotlib as mpl
from matplotlib import pyplot as plt
from pandas import DataFrame


# df_base_gr = df_base.groupby(by=["arrival_date_month"]).sum()
df_base_gr = df_base.groupby("arrival_date_month")[["arrival_date_month","arrival_date_day_of_month"]].sum()
df_base_gr
```

```python
plt.style.use('bmh')

df_plot = DataFrame(df_base_gr,columns=["arrival_date_day_of_month","arrival_date_month"])
df_plot.plot(x ="arrival_date_month", y="arrival_date_day_of_month", kind = 'bar')
plt.title("Most Visitor Come") 

plt.show()
```

```python
df_plot.plot.pie(y='arrival_date_day_of_month',figsize=(10, 10),autopct='%1.1f%%', startangle=90)
plt.title('Most Visitor Come')
plt.show()
```
