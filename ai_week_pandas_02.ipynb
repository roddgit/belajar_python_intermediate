1. Find all trips that is greater than 10 miles and originated from New York and Morrisville


```python
import pandas as pd
import numpy as np
datauber = pd.read_csv('csv/My Uber Drives - 2016.csv')
#print(datauber)

datagreater = datauber.loc[(datauber["MILES*"] > 10) & (datauber["START*"].isin(["New York","Morrisville"]))]

datagreater.head(n = 10)
```

 Exercise 5

 Create a new column with the following condition:
 a) >10 : Long Trip
 b) 5-10 : Medium Trip
 c) <5 : Short Trip

```python
datauber["TYPE_TRIP"] = np.where(datauber["MILES*"] > 10, "Long Trip",(np.where((datauber["MILES*"] >= 5) & (datauber["MILES*"] <= 10),"Medium Trip","Short Trip")))
datauber.head(n = 10)
```
